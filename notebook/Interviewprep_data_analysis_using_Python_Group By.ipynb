## Interviewprep data analysis using Python - Group By

```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv("data/fortune1000.csv", index_col = 'Rank')
fortune.head()
```

```python
sectors = fortune.groupby('Sector')
sectors
```

```python
# First Operations with groupby Object
sectors.size()
```

```python
sectors.first()
```

```python
sectors.groups
```

```python
# Retrieve a group from a GroupBy object with the get_group Method
sectors.get_group('Retailing')
```

```python
# Methods on the Groupby Object and DataFrame Columns
sectors.max()
```

```python
sectors.min()
```

```python
sectors.sum()
```

```python
sectors['Revenue'].mean()
```

```python
# The agg Method
sectors.agg({'Revenue': 'sum',
             'Profits': 'sum',
             'Employees': 'mean'})
```

```python
sectors.agg(['sum', 'mean', 'std'])
```

```python
# Iterating through Groups
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_company_in_group)
    
df   
```
